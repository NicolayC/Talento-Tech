**Medidas de distancia en datos**

Datos:

```python
import numpy as np
import pandas as pd
```

```python
x1 = int(input("x1: "))
y1 = int(input("y1: "))
x2 = int(input("x2: "))
y2 = int(input("y2: "))
```

**Distancia Euclidiana**

```python
def distanciaEuclidiana(x1, x2, y1, y2):
  distancia = np.sqrt(pow(x2 - x1, 2) + pow(y2 - y1, 2))
  return distancia
distanciaEuclidiana(x1, x2, y1, y2)
```

**Distancia Manhattan**

```python
def distanciaManhattan(x1, x2, y1, y2):
  distancia = abs(x2 - x1) + abs(y2 - y1)
  return distancia
distanciaManhattan(x1, x2, y1, y2)
```

**Distancia Hamming**

Código sugerido por ChatGPT

```python
valor_1 = input("Ingresa el primer valor: ")
valor_2 = input("Ingresa el segundo valor: ")

def distanciaHamming(valor_1, valor_2):
  distancia = 0
  for i in range (len(valor_1)):
    if valor_1[i] != valor_2[i]:
      distancia += 1

  return distancia
distanciaHamming(valor_1, valor_2)
```

**Distancia Chebyshev**

```python
def distanciaChebyshev(x1, x2, y1, y2):
  distancia = max(abs(x1 - x2), abs(y1 - y2))
  return distancia
distanciaChebyshev(x1, x2, y1, y2)
```

**Distancia Canberra**

```python
def distanciaCanberra(x1, x2):
    numerador = abs(x1 - x2)
    denominador = abs(x1 - abs(x2))
    distancia = numerador / denominador

    return distancia
distanciaCanberra(x1, x2)
```

**Distancia Coseno**

```python
def distanciaCoseno(x1, x2):
  distancia = (x1 * y1) / np.sqrt(pow(x1, 2) + pow(x2, 2))

  return distancia

distanciaCoseno(x1, x2)
```

**Distancia de Jaccard**
Código sugerido por ChatGPT

```python
def distanciaDeJaccard(x1, x2, y1, y2):
  set_x = set([x1, x2])
  set_y = set([y1, y2])
  interseccion = set_x & set_y
  union = set_x | set_y
  distancia = len(interseccion) / len(union)
  return distancia

distanciaDeJaccard(x1, x2, y1, y2)
```
