Importaciones

```python
import keras
import tensorflow as tf
```

Lista de frases

```python
frases = ['Hola mundo',
          'Hola a todos',
          'Hola a todo el mundo',
          'Buena día, como estas hoy']
```

Diccionario de tokens

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words = 10,
                                                  oov_token="<OOV>")
tokenizer.fit_on_texts(frases)
word_index = tokenizer.word_index
print('word_index =', word_index)
```

Generación de secuencias tokenizadas

```python
secuencias = tokenizer.texts_to_sequences(frases)
print('secuencias =', secuencias)
```

Llenado de secuencias a una longitud uniforme

```python
relleno = keras.preprocessing.sequence.pad_sequences(secuencias)
print('rellena =\n', relleno)
```

Truncamiento

```python
relleno = keras.preprocessing.sequence.pad_sequences(secuencias,
                                                     padding = 'post',
                                                     truncating = 'post')
print('rellena =\n', relleno)
```
