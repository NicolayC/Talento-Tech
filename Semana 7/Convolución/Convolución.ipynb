Prompt usado en ChatGPT: Chat, tengo estas dos listas x=[1, 2, 3, 4, 5 ] (posición 0 : 2),            h=[ -1, 5, 3, -2, 1 ] (posición 0: 3), ¿Cómo lo puedo hacer en python? además, como plotearía x, h y el resultado de la convolución de estos dos valores, y?

```python
import numpy as np

x = [1, 2, 3, 4, 5]
h = [-1, 5, 3, -2, 1]

# Realizar la convolución de x[0:3] y h[0:4] con mode='full'
y = np.convolve(x[0:3], h[0:4], mode='full')

print(y)
```

```python
import matplotlib.pyplot as plt

# Plotear las listas originales x y h
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.stem(x, use_line_collection=True)
plt.title('Lista x')

plt.subplot(3, 1, 2)
plt.stem(h, use_line_collection=True)
plt.title('Lista h')

# Plotear el resultado de la convolución y
plt.subplot(3, 1, 3)
plt.stem(y, use_line_collection=True)
plt.title('Resultado de la convolución')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Graficar x
plt.plot(n[:len(x)], x, marker='o', label='x[n]')

# Graficar h
plt.plot(n[:len(h)], h, marker='o', label='h[n]')

# Graficar y (resultado de la convolución)
plt.plot(n, y, marker='o', label='y[n] = x[n] * h[n]')

plt.title('Convolution')
plt.xlabel('n')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.show()
```
